```python
''' 
Right rotate an array by 1 steps
'''
```

```python
nums = [ 23, 45, 12, 78, 56, 89 ]
temp = nums[-1]
for i in range(len(nums)-1, 0, -1):
    nums[i] = nums[i-1]
nums[0] = temp
print(nums)
```

```python
#Right rotate an array by k steps
nums = [ 23, 45, 12, 78, 56, 89 ]
k = 3
k = k % len(nums)
temp = nums[-k:]
for i in range(len(nums)-1, k-1, -1):
    nums[i] = nums[i-k]
nums[:k] = temp
print(nums)

```
