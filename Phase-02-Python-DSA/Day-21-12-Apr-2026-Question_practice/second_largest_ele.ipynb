```python
'''
Find the second largest element 


'''
```

```python
nums =[56, 23, 89, 45, 12, 78]
nums.sort()
print(nums[-2])
#time complexity is O(nlogn) because of the sort function.
```

```python
# better approach
nums =[56, 23, 89, 45, 12, 78]
largest = float('-inf')
second_largest = float('-inf')
for num in nums:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num
print(second_largest)
#time complexity is O(n) because we are traversing the list only once.
```
