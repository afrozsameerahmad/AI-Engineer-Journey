```python
n = 100
k = 0

for i in range(n // 2, n + 1):
    j = 2
    while j <= n // 2:
        k = k + n // 2
        j *= 2

print(k)
```

## Complexity Analysis
- Outer loop runs from `n // 2` to `n`, so it runs about `n/2` times, i.e., `O(n)`.
- Inner loop doubles `j` each step (`j *= 2`) until `n // 2`, so it runs `O(log n)` times.
- Total complexity is `O(n log n)` as both loops multiply.
