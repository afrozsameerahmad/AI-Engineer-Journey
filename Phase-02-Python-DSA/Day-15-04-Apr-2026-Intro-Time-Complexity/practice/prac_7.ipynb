```python
def factorial(n):
    if n==1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))
```

```python
''' 
when recursive function calls itself, it creates a new stack frame for each call.
we count function calls to determine time complexity.
In this case, the function calls itself n times until it reaches the base case (n==1).
Therefore, the time complexity of this code is O(n) because the number of function calls grows linearly with the input n.
'''
```
