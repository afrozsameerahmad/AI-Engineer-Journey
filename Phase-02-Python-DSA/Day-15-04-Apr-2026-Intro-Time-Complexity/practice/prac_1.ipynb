```python
l = [1, 2, 3, 4, 5]
sum =0 
for i in l:
    sum+=i
print(sum)
```

```python
product =1
for i in l:
    product*=i
print(product)
```

```python
''' 
first program has 5 operations and the second program has 5 operations.
So the time complexity of both programs is O(n) where n is the size of the input list l.
O(n)+O(n) = O(2n) = O(n)
in simple when we have one loop then time complexity most of time is O(n)  

'''
```
