```python
def fib(n):
    if n==1 or n==0:
        return 1
    else:
        return fib(n-1) + fib(n-2)
print(fib(5))
```

```python
''' 

time complexity of above code is O(2^n) because
each call to fib(n) results in two additional calls to fib(n-1) and fib(n-2).
This creates a binary tree of function calls, where each level of the tree represents a call to fib with a smaller value of n.
The number of calls grows exponentially as n increases, leading to a time complexity of O(2^n). 
'''
```

```python
def fib(n):
    if n == 0 or n == 1:
        return 1
    
    a, b = 1, 1
    for _ in range(2, n+1):
        a, b = b, a + b
    
    return b

print(fib(5))
# better approach with O(n) time complexity and O(1) space complexity because we are using only two variables to store the last two Fibonacci numbers and iterating through the loop n times.
```
