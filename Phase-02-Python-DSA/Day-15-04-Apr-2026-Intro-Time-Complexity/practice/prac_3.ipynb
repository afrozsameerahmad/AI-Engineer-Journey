```python
def intostr(n):
    digits = '0123456789'
    if n==0:
        return '0'
    result = ''
    while n > 0:
        result = digits[n % 10] + result
        n //= 10    
    return result
print(intostr(12345))
print(intostr(0))
print(type(intostr(12345)))
```
