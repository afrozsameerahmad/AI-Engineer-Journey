```python
l =[1, 2, 3, 4, 5]
for i in range(0, len(l)):
    for j in range(i+1, len(l)):
        print('{}, {}'.format(l[i], l[j]))
```

```python
''' 
time complexity of above code is O(n^2) because 

outer loop runs n times and inner loop runs n-1 times in first iteration, n-2 times in second iteration and so on.
So total number of iterations will be n*(n-1)/2 which is O(n^2)
'''
```
