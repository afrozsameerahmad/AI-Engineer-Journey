```python
''' 
-----ORDER OF GROWTH


Order of growth is a way to classify algorithms based on how their running time or space requirements grow as the input size grows.
It is a way to compare the efficiency of different algorithms and to predict how they will perform as the input size increases.

'''

''' 
Measuring order of growth -- Big O notation

Big O notation measures an upper bound on the asymptotoc growth , often called order of growth

- Big Oh or O() is used to decribe the worst case time complexity of an algorithm. 
It gives an upper bound on the growth rate of the algorithmS running time as the input size increases.
- Big Omega or Ω() is used to describe the best case time complexity of an algorithm.
It gives a lower bound on the growth rate of the algorithm's running time as the input size increases.
- Big Theta or Θ() is used to describe the average case time complexity of an algorithm.    
It gives a tight bound on the growth rate of the algorithm's running time as the input size increases.

'''
```

```python
'''  
Types of time complexity
1- Constant time complexity O(1)
2- Logarithmic time complexity O(log n)
3- Linear time complexity O(n)
4- Linearithmic time complexity O(n log n)
5- Quadratic time complexity O(n^2)
6- Exponential time complexity O(2^n)


'''
```

```python
''' 
1- Constant time complexity O(1)

An algorithm is said to have constant time complexity if the time taken to execute does not depend on the size of the input.
This means that the algorithm will take the same amount of time to execute regardless of the size of the input.
Examples of constant time complexity include:
- Accessing an element in an array by index
- Checking if a number is even or odd



'''
```

```python
''' 
2-LINEAR TIME COMPLEXITY O(n)
An algorithm is said to have linear time complexity if the time taken to execute grows linearly with the size of the input.
This means that if the input size doubles, the time taken to execute will also double.
Examples of linear time complexity include:
- Iterating through an array
- Finding the maximum or minimum element in an array
- Linear search algorithm


'''
```

```python
''' 
3-logarithmic time complexity O(log n)
An algorithm is said to have logarithmic time complexity if the time taken to execute grows logarithmically with the size of the input.
This means that if the input size doubles, the time taken to execute will only increase by a  constant amount.
Examples of logarithmic time complexity include:
- Binary search algorithm
- Finding the height of a balanced binary search tree

'''
```

```python
''' 
4- Linearithmic time complexity O(n log n)
An algorithm is said to have linearithmic time complexity if the time taken to execute grows as the product of the input size and the logarithm of the input size.
This means that if the input size doubles, the time taken to execute will increase by a factor of 2 log 2 = 2.
Examples of linearithmic time complexity include:       
- Merge sort algorithm
- Quick sort algorithm

'''
```

```python
''' 
5- Quadratic time complexity O(n^2)
An algorithm is said to have quadratic time complexity if the time taken to execute grows as the square of the input size.
This means that if the input size doubles, the time taken to execute will increase by a factor of 4.
Examples of quadratic time complexity include:
- Bubble sort algorithm
- Insertion sort algorithm


'''
```

```python
''' 
6- Exponential time complexity O(2^n)
An algorithm is said to have exponential time complexity if the time taken to execute grows exponentially with the size of the input.
This means that if the input size doubles, the time taken to execute will increase by a factor of 2^2 = 4.
'''
```
