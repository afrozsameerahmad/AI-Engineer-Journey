```python
''' 
Parameterized and  Functional recursion

Parameterized recursion is a type of recursion where the recursive function takes one or more parameters that are used to control the flow of the recursion.

Functional recursion is a type of recursion where the recursive function returns a value that is used in the next recursive call.

sum of 1 to n [parameterized]
'''
```

```python
# sum of 1 to n  # parameterized recursion
def func(sum, i, n):
    if i > n:
        return sum
    return func(sum + i, i + 1, n)

print(func(0, 1, 5))
```

```python
# functional recursion
def func(n):
    if n == 1:
        return 1
    return n + func(n - 1)  
print(func(5))
```
