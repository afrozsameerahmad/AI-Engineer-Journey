```python
# Head Recursion  
# Head Recursion is when the recursive call happens first, and work is done after the call returns.
def func(n):
    if n == 0:
        return
    func(n-1)          
    print('sameer')   
func(5)
# time complexity is O(n) and space complexity is O(n) because of the call stack.
```

```python
# Tail Recursion (Definition)

# Tail Recursion is when the recursive call is the last statement, and no work is left after it.
def func(n):
    if n == 0:
        print('done')
        return
    print('sameer')     
    func(n-1)          

func(5)
# time complexity is O(n) and space complexity is O(n) because of the call stack.
```
