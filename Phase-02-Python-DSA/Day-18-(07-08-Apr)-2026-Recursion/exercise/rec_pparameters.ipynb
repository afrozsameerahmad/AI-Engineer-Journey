```python
''' 
Recursion using parameters is a powerful technique that allows a function to call itself with different arguments. 
This approach can simplify code and make it more elegant, especially when dealing with problems that have a natural recursive structure, 
such as tree traversals, combinatorial problems, and divide-and-conquer algorithms.


'''
```

```python
# print x ,n times using recursion
def func(x,n):
    if n == 0:
        return
    print(x)
    func(x,n-1)
func(15,4)
```

```python
# print 1 to n using recursion
# head recursion
def func(i,n):
    if i > n:
        return
    print(i)
    func(i+1,n)
func(1,5)
```

```python
# back tracking
def func(i,n):
    if i > n:
        return
    func(i+1,n)
    print(i)
func(1,5)
```
