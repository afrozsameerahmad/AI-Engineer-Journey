```python
''' 
Check if a  string is a palindrome using recursion and loop 

'''
```

```python
s = 'nitin'
l = 0
n = len(s)
r = n - 1

while l < r:
    if s[l] != s[r]:
        print('not a palindrome')
        break
    l += 1
    r -= 1
else:
    print('palindrome')
# time complexity O(n) and space complexity O(1)
```

```python
s='anbcddcbna'
n=len(s)
def func(s,l,r):
    if l >= r:
        return True
    if s[l] != s[r]:
        return False
    return func(s,l+1,r-1)
if func(s,0,n-1):
    print('palindrome')
else:  
    print('not a palindrome')
# time complexity O(n) and space complexity O(n) due to recursive stack
```
