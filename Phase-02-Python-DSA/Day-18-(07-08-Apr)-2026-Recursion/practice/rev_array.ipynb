```python
'''
Reverse an array using recursion
'''
```

```python
 
''' 
num=[5,7,3,2,6,1,5,9,2]
left =2 
right = 5
[5,7,1,6,2,3,5,9]

left = 0
right = 7
[9,5,1,6,2,3,7,5]
'''
```

```python
nums = [5,7,3,2,6,1,5,9,2]

def func(nums, left, right):
    if left >= right:
        return
    nums[left], nums[right] = nums[right], nums[left]
    func(nums, left + 1, right - 1)

def rev_array(nums, left, right):
    func(nums, left, right)
    return nums

print(rev_array(nums, 1, 5))
```
