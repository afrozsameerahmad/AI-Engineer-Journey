```python
''' 
Largest Consecutive Sequence

'''
nums=[1,99,101,98,2,5,3,100,1,1]
n=len(nums)
max_count=0
for i in range(0,n):
    num=nums[i]
    count=1
    while num + 1 in nums:
        num +=1
        count +=1
    max_count=max(max_count,count)  
print(max_count)
# time complexity is O(n^2) and space complexity is O(1)
```

```python
#optimal solution
nums=[1,99,101,98,2,5,3,100,1,1]
nums.sort()
n=len(nums)
last_smaller=float('-inf')
longest=0
count=0
for i in range(0,n):
    num = nums[i]
    if num -1 == last_smaller:
        count +=1
        last_smaller=num
    elif num != last_smaller:
        count=1
        last_smaller=num
    longest=max(longest,count)
print(longest)
# time complexity is O(nlogn) and space complexity is O(1)
```

```python
#more optimum solution
nums=[1,99,101,98,2,5,3,100,1,1]
n=len(nums)
num_set=set(nums)
longest=0
for i in range(0,n):
    num_set.add(nums[i])
longest=0
for num in num_set:
    if num -1 not in num_set:
        x=num
        count=1
        while x +1 in num_set:
            count +=1
            x+=1
        longest=max(longest,count)
print(longest)
# time complexity is O(n) and space complexity is O(n)
```
