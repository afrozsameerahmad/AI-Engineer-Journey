```python
nums=[1,1,1,2,3,3,5,6,7,7,7,9,12,12,13]
n=len(nums)
lb=-1
low=0
high=n-1
target=7
while low<=high:
    mid=low+(high-low)//2
    if nums[mid]>=target:
        lb=mid
        high=mid-1
    else:
        low=mid+1
print(lb)
#time complexity: O(logn)
#space complexity: O(1)
```

```python
nums=[1,1,1,2,3,3,5,6,7,7,7,9,12,12,13]
def lower_bound(nums,target):
    n=len(nums)
    lb=-1
    low=0
    high=n-1
    while low<=high:
        mid=low+(high-low)//2
        if nums[mid]>=target:
            lb=mid
            high=mid-1
        else:
            low=mid+1
    return lb
print(lower_bound(nums,7))
```
