```python
nums=[1,1,1,2,3,3,5,6,7,7,7,9,12,12,13]
n=len(nums)
ub=n
low=0
high=n-1
target=7

while low <= high:
    mid = low + (high - low) // 2
    
    if nums[mid] > target: 
        ub = mid
        high = mid - 1
    else:
        low = mid + 1

print(ub)
#time complexity: O(log n)
#space complexity: O(1)
```
