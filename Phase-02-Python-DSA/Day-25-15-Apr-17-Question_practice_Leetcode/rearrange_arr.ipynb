```python
''' 
Rearrange elements by sign


'''
```

```python
nums = [5,10,-2,-1,-10,6]

pos = []
neg = []
for num in nums:
    if num > 0:
        pos.append(num)
    else:
        neg.append(num)
for i in range(len(pos)):
    nums[2*i] = pos[i]
    nums[2*i + 1] = neg[i]

print(nums)
# time complexity O(n)
# space complexity O(n)
```

```python
#optimized approach
nums = [5,10,-2,-1,-10,6]
n=len(nums)
result=[0]*n
pos_index,neg_index=0,1
for i in range(0,n):
    if nums[i]>0:
        result[pos_index]=nums[i]
        pos_index+=2
    else:
        result[neg_index]=nums[i]
        neg_index+=2
print(result)
# time complexity O(n)
# space complexity O(n)
```
