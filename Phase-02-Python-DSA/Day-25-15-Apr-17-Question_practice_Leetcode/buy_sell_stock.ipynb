```python
'''  
Stock Buy and Sell
 maximum profit that can be achieved by buying and selling a stock at most once.
'''
#brute force 
prices=[7,2,5,1,6,4,8]
n=len(prices)
max_profit=0
for i in range(0,n):
    for j in range(i+1,n):
        if prices[j]>prices[i]:
            p=prices[j]-prices[i]
            max_profit=max(max_profit,p)
print(max_profit)   
# time complexity O(n^2)
# space complexity O(1)
```

```python
prices=[7,2,5,1,6,4,8]
n=len(prices)
max_profit=0
min_price=float('inf')
for i in range(0,n):
    min_price=min(min_price,prices[i])
    profit=prices[i]-min_price
    max_profit=max(max_profit,profit)
print(max_profit)
# time complexity O(n)
# space complexity O(1)
```
