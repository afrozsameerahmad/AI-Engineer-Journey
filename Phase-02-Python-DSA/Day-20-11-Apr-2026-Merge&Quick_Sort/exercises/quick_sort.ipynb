```python
'''
nums=[4,1,7,6,3,2,8]

1> Pick a pivot
- it can be the first element, last element, middle element or any random element.
- In this example, we will pick the last element as the pivot.
2> Partitioning
- We will partition the array into two sub-arrays based on the pivot.
- Elements less than the pivot will go to the left sub-array, and elements greater than the pivot will go to the right sub-array.
- After partitioning, we will have:
- Left sub-array: [4, 1, 3, 2]
- Pivot: 8
- Right sub-array: [7, 6]
3> Recursively apply the same process to the left and right sub-arrays until we reach base cases (sub-arrays of size 0 or 1).
- For the left sub-array [4, 1, 3, 2]:
- Pick a pivot (let's say 2).
- Partitioning:
- Left sub-array: [1]
- Pivot: 2
- Right sub-array: [4, 3]
- For the right sub-array [7, 6]:
- Pick a pivot (let's say 6).
- Partitioning:
- Left sub-array: []
- Pivot: 6
- Right sub-array: [7]
4> Combine the sorted sub-arrays and pivots to get the final sorted array.
- After sorting the left sub-array [1, 2, 3, 4]
- After sorting the right sub-array [6, 7]
- Final sorted array: [1, 2, 3, 4, 6, 7, 8]



'''
```

```python
nums=[4,1,7,6,3,2,8]
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[-1]
    left = [x for x in arr[:-1] if x < pivot]
    right = [x for x in arr[:-1] if x >= pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)
sorted_nums = quick_sort(nums)
print(sorted_nums)
# time complexity: O(n log n) on average, O(n^2) in the worst case when the smallest or largest element is always chosen as the pivot.
# space complexity: O(n) due to the additional space used for the left and right sub-arrays during the recursive calls.
```

```python
# quick sort recursive

nums=[4,1,7,6,3,2,8]
def partition(nums, low, high):
    pivot = nums[high]
    i = low - 1
    for j in range(low, high):
        if nums[j] < pivot:
            i += 1
            nums[i], nums[j] = nums[j], nums[i]
    nums[i + 1], nums[high] = nums[high], nums[i + 1]
    return i + 1    

def quick_sort(nums,low,high):
    if low < high:
        pivot_index = partition(nums, low, high)
        quick_sort(nums, low, pivot_index - 1)
        quick_sort(nums, pivot_index + 1, high)
quick_sort(nums, 0, len(nums) - 1)
print(nums)
```
