```python
'''
Binary  Search

1-A real life example where binary search work
2-Coding problem
3- iterative solution
4-Recursive solution
5-time and space complexity

'''
```

```python
nums=[2,4,6,7,9,11,18,19]
target=6
def binary_search(nums,target):
    low =0
    high=len(nums)-1
    while low<=high:
        mid=low+(high-low)//2
        if nums[mid]==target:
            return mid
        elif nums[mid]<target:
            low=mid+1
            
        else:
            high=mid-1
    return -1
print(binary_search(nums,target))
#time complexity is O(log n) and space complexity is O(1) for iterative solution

```

```python
# recursive solution
nums=[2,4,6,7,9,11,18,19]
target=11
def binary_search(nums,target,low,high):
    if low >high:
        return -1
    mid=low+(high-low)//2
    
    if nums[mid]==target:
        return mid
    elif nums[mid]<target:
        return binary_search(nums,target,mid+1,high)
    else:
        return binary_search(nums,target,low,mid-1)
print(binary_search(nums, target, 0, len(nums) - 1))
# time complexity is O(log n) and space complexity is O(1)  
```
