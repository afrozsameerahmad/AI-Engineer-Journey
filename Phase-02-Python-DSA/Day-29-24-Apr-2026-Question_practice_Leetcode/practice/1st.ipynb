```python
# practice question
nums=[-1,0,3,5,9,12]
target=9
def binary_search(nums,target,low,high):
    if low>high:
        return -1
    mid=low+(high-low)//2
    if nums[mid]==target:
        return mid
    elif nums[mid]<target:
        return binary_search(nums,target, mid+1,high)
    else:
        return binary_search(nums,target,high,mid-1)
print(binary_search(nums,target,0,len(nums)-1))
```
