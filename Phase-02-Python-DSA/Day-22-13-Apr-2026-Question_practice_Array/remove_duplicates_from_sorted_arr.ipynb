```python
''' 
remove duplicates from a sorted array 
in place 7

'''
```

```python
#brute force
nums = [1,2,3,4,5,7,9,11,11,12,34,34,56,56,56]

unique = [nums[0]]

for i in range(1, len(nums)):
    if nums[i] != nums[i-1]:
        unique.append(nums[i])

print(len(unique))
print(unique)
#time complexity is O(n) and space complexity is O(n) because we are using an extra list to store unique elements.
```

```python
#optimal solution two pointer approach
nums=[1,1,1,2,3,4,4,7,9,9,9,10]
n= len(nums)
if n == 1:
    print(1)
i=0
j=i+1
while j < n:
    if nums[j]!= nums[i]:
        i+=1
        nums[i]= nums[j]
    j+=1
print(i+1)
#time complexity is O(n) 
# space complexity is O(1)  
```
