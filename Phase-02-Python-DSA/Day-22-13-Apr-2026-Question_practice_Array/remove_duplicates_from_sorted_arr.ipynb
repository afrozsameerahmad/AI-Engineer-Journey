```python
''' 
remove duplicates from a sorted array 
in place 7

'''
```

```python
nums = [1,2,3,4,5,7,9,11,11,12,34,34,56,56,56]

unique = [nums[0]]

for i in range(1, len(nums)):
    if nums[i] != nums[i-1]:
        unique.append(nums[i])

print(len(unique))
print(unique)
#time complexity is O(n) and space complexity is O(n) because we are using an extra list to store unique elements.
```
