```python
nums = [3,5,6,7,8,9,12,15,18,21,24,27,30]
n=len(nums)
is_sorted=True
for i in range(0,n-1):
    if nums[i]>nums[i+1]:
         is_sorted=False
         break
if is_sorted:
    print("sorted")
else:    print("not sorted")
#time complexity is O(n) because we have to check each element once.
#space complexity is O(1) because we are using only a constant amount of extra space to store the boolean variable is_sorted.
```
