```python
'''
1- Linked List is linear data structure where each element is a separate object. 
Each element (node) of a list is comprising of two items - the data and a reference to the next node.
The last node has a reference to null. 
The entry point into a linked list is called the head of the list. 
It should be noted that head is not a separate node, but the reference to the first node. 
If the list is empty then the head is a null reference.

2-Where to use Linked List?
Linked List is used when we have to perform many insertions and deletions in the list.
as compared to arrays, linked list is dynamic in nature and the size of the linked list can grow or shrink as needed.

3- Types of Linked List:
1- Singly Linked List: In this type of linked list, each node contains data and a reference to the next node in the list. It allows for traversal in only one direction.
2- Doubly Linked List: In this type of linked list, each node contains data and two references: one to the next node and another to the previous node. It allows for traversal in both directions.
3- Circular Linked List: In this type of linked list, the last node points back to  the first node, creating a circular structure. It can be either singly or doubly linked.    

example where it used in real life:
1- Music Player: A music player can use a linked list to manage the playlist. Each song can be represented as a node in the linked list, allowing for easy addition and removal of songs from the playlist. 
2- Web Browser: A web browser can use a linked list to manage the browsing history. Each webpage visited can be represented as a node in the linked list, allowing for easy navigation through the history and efficient memory usage.  


'''
```
