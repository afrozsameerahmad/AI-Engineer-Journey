```python
''' Maximum Subarray '''


nums = [-2,1,-3,4,-1,2,1,-5,4]

n = len(nums)
maxi = float('-inf')

for i in range(n):
    total = 0
    for j in range(i, n):
        total += nums[j]
        maxi = max(maxi, total)

print(maxi)
#time complexity: O(n^2)    
# space complexity: O(1)
```

```python
#using kadane's algorithm

nums = [-2,1,-3,4,-1,2,1,-5,4]

n = len(nums)
maxi = float('-inf')
total = 0
for i in range(0,n):
    total += nums[i]
    maxi = max(maxi, total)
    if total < 0:
        total = 0
print(maxi)
#time complexity: O(n)
# space complexity: O(1)
```
