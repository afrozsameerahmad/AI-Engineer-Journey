```python
''' 
Hashing in python

Prestoring values into some data structure like list/dict/sets and then fethcing it .

'''
```

```python
''' 
Contraints 
1- 1<=n[i]<=10
2-n can have 10^8 elements 
2-m can have 10^8 elements 

'''


# brute force 
n=[5,3,2,2,1,5,5,7,5,10]
m=[10,111,1,9,5,67,2]
for num in m:
    count =0
    for x in n:
        if x==num:
            count+=1
    print(f"{num}: {count}")
# time complexity O(n*m)
#space O(1)

''' 
TC -> O(m*n)
10^8 * 10^8 = 10^16

TLE - TIME LIMIT EXCEEDED


'''
```

```python
# Optimal way 
hash_list=[0]*11
for i in n:
    hash_list[i]+=1
    
for i in m :
    if i<1 or i>10:
        print(0)
    else:
        print(f'{i}: {hash_list[i]}')
# time complexity O(n+m)
# space complexity O(1) since we are using constant space of 11 for hash_list
```

```python
#using dictionary
n=[5,3,2,2,1,5,5,7,5,10]
m=[10,111,1,9,5,67,2]
hash_dict={}
for num in n:
    hash_dict[num]=hash_dict.get(num,0)+1
    
for num in m:
    print(f'{num}: {hash_dict.get(num,0)}')
    
    
 
```

```python
# character hashing
''' 


'''
s='azyxyyzaaaa'
q=['d','a','y','x']
freq={}
for char in s:
    freq[char]=freq.get(char,0)+1
for char in q:
    print(f'{char} : {freq.get(char,0)}')
#time complexity O(n+q) where n is the length of string and q is the number of queries
# space complexity O(k) where k is the number of unique characters in the string
```
