```python
'''

Input:

arr = [4,5,1,2,0,4]

Output:

5
 
'''
#First Non-Repeating Element
arr = [4,5,1,2,0,4]
freq = {}
for i in arr:
    freq[i] = freq.get(i, 0) + 1
for i in arr:
    if freq[i] == 1:
        print(i)
        break
    
```

```python
#First Repeating Element
arr = [10,5,3,4,3,5,6]
freq = {}
for i in arr:
    freq[i]=freq.get(i,0)+1
for i in arr:
    if freq[i]>1:
        print(i)
        break
    
```
