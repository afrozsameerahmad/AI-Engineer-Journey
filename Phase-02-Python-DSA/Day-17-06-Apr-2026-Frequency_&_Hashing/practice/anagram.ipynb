```python
''' 

. Check Anagram

Input:

s1 = "listen"
s2 = "silent"

Output:

True

'''
s1 = "listen"
s2 = "silent"
print(sorted(s1) == sorted(s2))
```

```python
s1 = "listen"
s2 = "silent"

freq = {}

for ch in s1:
    freq[ch] = freq.get(ch, 0) + 1

for ch in s2:
    freq[ch] = freq.get(ch, 0) - 1

print(all(v == 0 for v in freq.values()))
```

```python
''' 

Input:

a = [1,2,2,1]
b = [2,2]

➡️ Output:

[2,2]
'''
a = [1,2,2,1]
b = [2,2]
freq = {}


for i in a:
    freq[i] = freq.get(i, 0) + 1
result = []
for i in b:
    if freq.get(i, 0) > 0:
        result.append(i)
        freq[i] -= 1
print(result)
```

```python
''' 
Input:

a = [1,2,3]
b = [2,3,4]

 Output:

[1,2,3,4]

'''
a = [1,2,3]
b = [2,3,4]
freq = {}
for i in a:
    freq[i] = freq.get(i, 0) + 1
for i in b:
    freq[i] = freq.get(i, 0) + 1
print(list(freq.keys()))
```
