```python
'''
input 
s = "banana"

output
{'b':1, 'a':3, 'n':2}


'''
#brute force approach
s = "banana"
freq={}
for i in range(0,len(s)):
    if s[i] in freq:
        freq [s[i]]+=1
    else:
        freq [s[i]]=1
print(freq)
```

```python
# best approach
s = "banana"
hash_map = dict()  # {}
for i in s:
    hash_map[i]= hash_map.get(i,0)+1
print(hash_map)
```
