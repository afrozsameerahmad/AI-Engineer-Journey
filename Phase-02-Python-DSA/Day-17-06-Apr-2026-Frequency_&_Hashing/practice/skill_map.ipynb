```python
''' 
Skill Matching System

Given:

student_skills = ["python","ml","sql","python","ml"]
query = ["python","java","ml"]

Output:

{"python":2, "java":0, "ml":2}
'''
```

```python

student_skills = ["python","ml","sql","python","ml"]
query = ["python","java","ml"]
freq={}
for i in student_skills:
    freq[i]=freq.get(i,0)+1
result={}
for q in query:
    result[q]=freq.get(q,0)
print(result)
```
