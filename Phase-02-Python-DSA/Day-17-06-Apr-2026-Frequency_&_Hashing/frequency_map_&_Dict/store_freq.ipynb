```python
''' 
Frequency map /Dictionary

Store frequency in dictionary

'''

nums=[5,6,7,7,1,9,111,1,1,5,1,1]
#how many times each number appears in the list
```

```python
#method 1
nums=[5,6,7,7,1,9,111,1,1,5,1,1]
freq_map = {}
for i in range(0,len(nums)):
    if nums[i] in freq_map:
        freq_map[nums[i]] +=1
    else:
        freq_map[nums[i]]=1
        
print(freq_map)
#time complexity O(n)
#space complexity O(n)
```

```python
# method 2
nums=[5,6,7,7,1,9,111,1,1,5,1,1]
n= len(nums)
hash_map=dict()
for i in range(0,n):
    hash_map[nums[i]]=hash_map.get(nums[i],0)+1
print(hash_map)
```
