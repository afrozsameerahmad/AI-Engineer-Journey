```python
''' 
Search Insert Position
If target not found, return where it should be inserted
Example:

nums = [1,3,5,6], target = 2 → output = 1


'''
```

```python
nums = [1,3,5,6], 
target = 2
def search_insert(nums,target,low,high):
    if low>high:
        return -1
    mid=low+(high-low)//2
    if nums[mid]==target:
        return mid
    elif nums[mid]>target:
        return search_insert(nums,target,low,mid-1)
    else:
        return search_insert(nums,target,mid+1,high)
    return low

print(search_insert(nums,target,0,len(nums)-1))
```
