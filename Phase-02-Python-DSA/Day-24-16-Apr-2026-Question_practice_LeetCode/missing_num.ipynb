```python
''' 
Find the missing number 
nums=[1,0,3,5] n=4  /2 output:2
nums=[9,6,4,3,2,5,7,0,1] n=10 /8

'''
```

```python
nums=[9,6,4,3,2,5,7,0,1]
n=len(nums)
for i in range(0,n+1):
    if i not in nums:
        print(i)
        
# time complexity O(n^2) and space complexity O(1)
        
```

```python
#using dictionary
nums=[9,6,4,3,2,5,7,0,1]
n=len(nums)
d={}
for i in range(0,n+1):
    d[i]=0
for i in nums:
    d[i]=1
for i in d:
    if d[i]==0:
        print(i)
# time complexity O(n) and space complexity O(n)
```

```python
nums=[9,6,4,3,2,5,7,0,1]
n=10
total_sum=n*(n-1)//2
actual_sum=sum(nums)
missing_num=total_sum-actual_sum
print(missing_num)
# time complexity O(n) and space complexity O(1)
```
