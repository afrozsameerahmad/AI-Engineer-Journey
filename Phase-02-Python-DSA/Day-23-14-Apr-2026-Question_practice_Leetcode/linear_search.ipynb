```python
''' 
Linear search 
Linear search is a simple algorithm that checks each element in a list sequentially until it finds the target value or reaches the end of the list. 
It is straightforward and does not require the list to be sorted.

'''
```

```python
nums=[1,0,2,4,3,0,0,5,1]  #target=5
n=len(nums)
target = 5
for i in range(n):
    if nums[i]==target:
        print(i)
        break
print(-1)
# time complexity: O(n)
# space complexity: O(1)
```
