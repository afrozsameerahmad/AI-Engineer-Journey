```python
''' 
Ceil The Floor

- Ceil = smallest no. in array >= target
- Floor = largest no. in array <= target
'''
```

```python
nums =[3,4,4,4,8,9,9,10,12,12,14,15]
target = 6
floor = -1
ceil = -1
low = 0
high = len(nums)-1
while low <= high:
    mid = low + (high-low)//2
    if nums[mid] == target:
        floor = nums[mid]
        ceil = nums[mid]
        break
    elif nums[mid] > target:
        ceil = nums[mid]
        high = mid - 1
    else:
        floor = nums[mid]
        low = mid + 1
print(floor,ceil)
# time complexity = O(logn)
# space complexity = O(1)
```
