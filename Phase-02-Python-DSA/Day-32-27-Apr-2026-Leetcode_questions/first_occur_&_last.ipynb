```python
#brute force approach
nums=[1,2,3,3,3,3,3,5,6,8,9,9,10]
target=3
first=-1
last=-1
n=len(nums)
for i in range(n):
    if nums[i]==target:
        if first==-1:
            first=i
        last=i
print(first,last)
```

```python
def find_bound(nums, target, is_first):
    low, high = 0, len(nums)-1
    ans = -1
    while low <= high:
        mid = (low + high) // 2
        if nums[mid] == target:
            ans = mid
            if is_first:
                high = mid - 1
            else:
                low = mid + 1
        elif nums[mid] > target:
            high = mid - 1
        else:
            low = mid + 1
    return ans

first = find_bound(nums, target, True)
last = find_bound(nums, target, False)
print(first, last)
```
