```python
# brute force approach
nums=[11,15,20,1,4,5,6,8,9,10]
target=8
n=len(nums)
for i in range(0,n):
    if nums[i]==target:
        print(i)
# time complexity: O(n)
#space complexity: O(1)
```

```python
#optimized approach using binary search
''' 
1- Identify sorted part
2- check if target is in that sorted part
3- if yes, then do binary search in that part

'''
nums = [17,18,20,1,3,4,5,7,8,10,11,13,14,16]
target = 4
def search_rotated(nums, target):
    low, high = 0, len(nums) - 1
    
    while low <= high:
        mid = low + (high - low) // 2
    
        if nums[mid] == target:
            return mid
        if nums[low] <= nums[mid]:
            if nums[low] <= target < nums[mid]:
                high = mid - 1
            else:
                low = mid + 1

        else:
            if nums[mid] < target <= nums[high]:
                low = mid + 1
            else:
                high = mid - 1
    
    return -1 

result = search_rotated(nums, target)
print(result)
```
