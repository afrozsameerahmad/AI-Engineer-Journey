```python
''' 
Armstrong number

153 = 1^3+5^3+3^3
    = 1+125+27
    =153

'''
```

```python
n=153
num =n
total = 0
nod=len(str(n))
while num>0:
    ld=num%10
    total = total +(ld **nod)
    num  //=10

print('armstrong' if total==n else'not armstrong')
#time complexity O(logn)
#space complexity O(1)
```
