```python
''' 
Insertion Sort Algorithm
It is a simple sorting algorithm that builds the final sorted array one item at a time.
It works by repeatedly taking the next unsorted element and inserting it into 
the correct position in the already sorted part of the array. 



'''
```

```python
nums=[12, 11, 13, 5, 6]
n=len(nums)
for i in range(1,n):
    key=nums[i]
    j=i-1
    while j>=0 and key<nums[j]:
        nums[j+1]=nums[j]
        j-=1
    nums[j+1]=key
print("Sorted array is:",nums)
#ti
#time complexity: O(n) in best case when the array is already sorted and O(n^2) in worst case.
#space complexity: O(1) as it is an in place sorting algorithm.
```
