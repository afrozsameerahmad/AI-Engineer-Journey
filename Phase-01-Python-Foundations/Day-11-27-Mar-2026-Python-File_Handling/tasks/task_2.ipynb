```python
''' 
Read through a text file line by line .use a dict to keep track of how many times each
vowel appears in the file. Finally print the resulting tabulation -- dictionary of vowel counts.

'''
def count_vowels(filename):
    vowels = 'aeiouAEIOU'
    vowel_count ={vowel : 0 for vowel in vowels}
    
    with open(filename,'r') as f:
        for line in f:
            for char in line:
                if char in vowels:
                    vowel_count[char] += 1
    return vowel_count

```

```python
print(count_vowels("sample.txt"))
```
