```python
''' 
Write a function which take two arguments:
the name of input file (to be read form)
and the output file  (which will be created)

for example if a file look like
abc  def
ghi  jkl

then the output file should look like
fed  cba
lkj  ihg
'''
def transform_file(input_file, output_file):
    with open(input_file, 'r') as i, open(output_file, 'w') as o:
        for line in i:
            parts = line.strip().split()   
            
            if len(parts) < 2:
                continue   
            word1 = parts[0]
            word2 = parts[1]
            
            
            rev1 = word1[::-1]
            rev2 = word2[::-1]
            
            
            o.write(rev2 + " " + rev1 + "\n")


 
transform_file("input.txt", "output.txt")
        
    

        
```
