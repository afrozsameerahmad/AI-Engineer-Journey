```python
''' 
Create a text file(using an editor not neccessarily Python)  containing two tab seprated 
columns with each column containing a number then use python to read through the file 
you have created  for each line multiply each fisrt number by the second and incluse it in the file in third column . in last add a line 
total by summing the value of third column 

'''


def process_file(filename):
    data = []
    total = 0

 
    with open(filename, 'r') as f:
        for line in f:
            parts = line.strip().split()    
            
             
            num1 = int(parts[0])
            num2 = int(parts[1])
            
           
            result = num1 * num2
            total += result
            
             
            data.append([str(num1), str(num2), str(result)])

    
    with open(filename, 'w') as f:
        for row in data:
            f.write('\t'.join(row) + '\n')
        
         
        f.write(f"Total\t{total}")

 
process_file("numbers.txt")
    

    
```
