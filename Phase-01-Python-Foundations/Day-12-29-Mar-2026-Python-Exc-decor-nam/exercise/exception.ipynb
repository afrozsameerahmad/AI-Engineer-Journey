### Exceptions

If things go wrong during the execution of the program(runtime). It generally happens when something unforeseen has happened.

- Exceptions are raised by python runtime
- You have to takle is on the fly

#### **Examples**

- Memory overflow
- Divide by 0 -> logical error
- Database error

```python
# Why is it important to handle exceptions
# how to handle exceptions
# -> Try except block
```

```python
# lets create a file
with open("sample.txt", "w") as f:
    f.write("This is a sample file.")
```

```python
#try catch block to handle exceptions
try:
     with open("sample1.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File not found.")
```

```python
# catching specific exceptions
try:
    f=open("sample.txt", "r")
    print(f.read())
    print(m)
except:
    print("An error occurred.")
```

```python
# how to catch specific error
try:
    m=5
    f=open("sample.txt", "r")
    print(f.read())
    print(m)
    print(5/2)
    l=[1, 2, 3]
    print(l[5])
except FileNotFoundError:
    print("File not found.")    
except NameError:
    print("Variable m is not defined.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
    
except Exception as e:
    print("An error occurred:", str(e))
    
    
```

```python
#else block in try except
try:
    f=open("sample.txt", "r")
except FileNotFoundError:
    print("File not found.")
except Exception:
    print("kuchh to lfda hai")
else:
    print(f.read())
```

```python
# finally block 
try:
    f=open("sample1.txt", "r")
except FileNotFoundError:
    print("File not found.")
except Exception:
    print("kuchh to lfda hai")
else:
    print(f.read())
finally:
    print("This will always execute.")

```

```python
# whats the importance of raise
# we can raise a error and except will catch that error and we can handle that error in except block
# example
class Bank:
    def __init__(self,balance):
        self.balance = balance
        
        
    def withdraw(self, amount):
        if amount <= 0:
            raise Exception("Withdrawal amount must be positive.")
        if amount>self.balance:
            raise Exception("Itne paise nahi hai tere  account me")
        else:
            self.balance -= amount
            print(f"Hogya Withdraw moz kr jake . Remaining balance: {self.balance}")
my_bank = Bank(1000)
try:
    my_bank.withdraw(6100)
except Exception as e:
    print(str(e))

 
```

```python
class SecurityError(Exception):
    def __init__(self, message):
        print(message)
    def logout(self):
        print('logout')

class Google:
    def __init__(self,name,email,password,device):
        self.name = name
        self.email = email
        self.password = password
        self.device = device
    
    def login(self,email,password,device):
        if device != self.device:
            raise SecurityError("Bhai teri to lag gyi")
        if email != self.email or password != self.password:
            raise SecurityError("Invalid email or password.")
        else:
            print("Login successful.")
    
obj=Google("sameer","sameer@example.com","password123","iPhone")

try:
    obj.login("sameer@example.com", "password123", "android")
except SecurityError as e:
    e.logout()
else:
    print(obj.name)
finally:
    print("database connection closed")
```
