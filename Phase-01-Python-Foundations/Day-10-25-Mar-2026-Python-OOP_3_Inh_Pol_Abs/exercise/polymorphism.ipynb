# Polymorphism means multiple faces
- Method Overloading
- Method overriding
- operator overloading

```python
# Method Overloading in Python it does not support method overloading but we can achieve it by using default arguments or variable length arguments.
class Shape:
    def area(self, a, b=None):
        if b is None:
            return a * a
        else:
            return a * b
    
s=Shape()
print(s.area(5)) 

print(s.area(5,10))
```

```python
# Method overiding in Python it is a feature that allows a subclass to provide a specific implementation of a method that is already defined in its superclass. The implementation in the subclass overrides the implementation in the superclass.
class Animal:
    def sound(self):
        return "Animal makes a sound"   
class Dog(Animal):
    def sound(self):
        return "Dog barks"
class Cat(Animal):
    def sound(self):
        return "Cat meows"
a=Animal()
d=Dog()
c=Cat()
print(a.sound())
print(d.sound())
print(c.sound())
```

```python
#operator overloading in Python it is a feature that allows us to define the behavior of operators for user-defined classes. We can overload operators by defining special methods in our class.
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __add__(self,other):
        return Point(self.x + other.x, self.y + other.y)
p1=Point(1,2)
p2=Point(3,4)
p3=p1+p2
print(p3.x,p3.y)
```
