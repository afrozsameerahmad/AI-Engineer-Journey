### Aggregation(Has-A relationship)

```python
'''  
Aggregation means has a relationship between two classes which means one class owns the other class but the owned class can exist without the owner class. 
It is a special form of association where one class contains a reference to another class. The contained class can exist independently of the container class.


# Example Customer has a address but the address can exist without the customer.

'''
```

```python
class Customer:
    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender
        self.address = address  
        
    
    def print_address(self):
        print(f"City: {self.address.get_city()}, Pin: {self.address.pin}, State: {self.address.state}")
        
    def edit_profile(self,new_name,new_city,new_pin,new_state):
        self.name = new_name
        self.address.edit_address(new_city,new_pin,new_state)
        
        
        
class Address:
    def __init__(self,city,pin,state):
        self.__city = city
        self.pin = pin
        self.state = state
    def get_city(self):
        return self.__city
    def edit_address(self,new_city,new_pin,new_state):
        self.__city = new_city
        self.pin = new_pin
        self.state = new_state
        
        
        
        
addr = Address("Lucknow",10001,"UP")
cust = Customer("Sameer","Male",addr)
cust.print_address()    
cust.edit_profile("Sameer Ahmad","Kanpur",20001,"UP")
cust.print_address()
```

```python
# Super Keyword is used to call the parent class constructor and methods in the child class. 
# It is used to access the parent class members (attributes and methods) from the child class.
class   Phone:
    def __init__(self,price ,brand,camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
    def buy(self):
        print("Buying a phone")
        
class SmartPhone(Phone):
    def buy(self):
        print("Buying a smartphone")
        super().buy()
s= SmartPhone(10000,"Samsung","48MP")   
s.buy()
```
