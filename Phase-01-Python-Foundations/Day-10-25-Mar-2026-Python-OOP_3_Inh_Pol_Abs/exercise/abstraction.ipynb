```python
''' 
Abstraction: Hiding the internal details and showing only the functionality to the user. 
It is achieved by using abstract classes and interfaces.


Example Bank App
1-WEB APPLICATION
2-MOBILE APPLICATION
whatever application we are using we can perform the same operations like check balance, withdraw money, deposit money etc.
but the internal implementation is different for each application. This is abstraction.

'''


```

```python
'''  
ABSTRACT CLASS: A class that cannot be instantiated and is meant to be subclassed. 
It can contain abstract methods that must be implemented by the subclasses.
'''
from abc import ABC, abstractmethod
class BankAccount(ABC):
    @abstractmethod
    def check_balance(self):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass

    @abstractmethod
    def deposit(self, amount):
        pass
class WebBankAccount(BankAccount):
    def __init__(self, balance):
        self.balance = balance

    def check_balance(self):
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance: {self.balance}")
class MobileBankAccount(BankAccount):
    def __init__(self, balance):
        self.balance = balance

    def check_balance(self):
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance: {self.balance}")
web_account = WebBankAccount(1000)
mobile_account = MobileBankAccount(500)
print("Web Account Balance:", web_account.check_balance())
print("Mobile Account Balance:", mobile_account.check_balance())
web_account.withdraw(200)
mobile_account.deposit(300)
```
