### Inheritance

```python
''' 
# Inheritance in Python is a fundamental object-oriented programming concept that allows a new class (called a child or subclass) to inherit attributes and methods from an existing class (called a parent or superclass).
# This promotes code reusability and establishes a natural hierarchical relationship between classes.

In simple terms the child class can access and use the properties and behaviors of parent class 
it make our code reusable and organized.


Example: Udemy website

-Student  :- Login ,register, view courses, enroll in courses
-Teacher :- Login, register, create courses, manage courses
-Admin :- Login, register, manage users, manage courses, view reports

in this example we can see that all three classes (Student, Teacher, Admin) have some common functionalities like login and register.
instead of writing the same code for login and register in each class we can create a parent class called User and put the common functionalities 
there and then the child classes (Student, Teacher, Admin) can inherit from the User class.
''' 
```

```python
# EXAMPLE 
class User:
    def __init__(self):
        self.name ='Sameer'
        
    def login(self):
        print(f'{self.name} logged in successfully')
        
class Student(User):
    def __init__(self):
        super().__init__()
        self.roll_number = 1
        
    def enroll(self):
        print(f'{self.roll_number} enrolled in course successfully')
        
u = User()
s=Student()
print(s.name)
print(s.roll_number)
s.login()
s.enroll()
```

```python
# Super Keyword is used to call the parent class constructor and methods in the child class. 
# It is used to access the parent class members (attributes and methods) from the child class.
# Example of super keyword

class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand=brand
        self.camera= camera
    def buy(self):
        print("Buying a phone")
        
class SmartPhone(Phone):
    def buy(self):
        print("Buying a smartphone")
        super().buy()
        
s=SmartPhone(20000,'Apple',13)
s.buy()
```

 #  Inheritance in summary
 - A class can inherit from another class
 - Inheritance improve code reuse 
 - Constructor ,attributes,methods get inherited to the child class
 - the parent has no access to the child class 
 - Private properties of parent class directly not accessed by child class we can access by using super keyword
 - Child class can override the parent attributes or methods of parent class . this is called method overloading
 - super() is im built function which is used to invoke the parent class methods and constructor

```python
class Parent:
    def __init__(self,num):
        self.__num=num
    def get_num(self):
        return self.__num
class Child(Parent):
    def __init__(self,num,val):
        super().__init__(num)
        self.__val=val
    def get_val(self):
        return self.__val
    
son=Child(10,20)
print(son.get_num())
print(son.get_val())
        
```

#  Types of INHERITANCE
- Single
- Multilevel
- Hierarchical
- Multiple (Diamond Problem)
- Hybrid 

```python
# Single Inheritance
# In single inheritance, a child class inherits from a single parent class.
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand= brand
        self.camera=camera 
    def buy(self):
        print("Buying a phone")
        
class SmartPhone(Phone):
    def __init__(self,price,brand,camera,os,ram):
        super().__init__(price,brand,camera)
        self.os=os
        self.ram= ram
        print("Inside the Smartphone constructor")
    def buy(self):
        print("Buying a smartphone")
        super().buy()
s=SmartPhone(20000,"Samsung",12,"Android",4)
s.buy()
```

```python
# Multilevel Inheritance
# In multilevel inheritance, a child class inherits from a parent class, and then another child class inherits from that child class, creating a chain of inheritance.
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand= brand
        self.camera=camera 
    def buy(self):
        print("Buying a phone")
        
class SmartPhone(Phone):
    def __init__(self,price,brand,camera,os,ram):
        super().__init__(price,brand,camera)
        self.os=os
        self.ram= ram
        print("Inside the Smartphone constructor")
    def buy(self):
        print("Buying a smartphone")
        super().buy()
class AndroidPhone(SmartPhone):
    def __init__(self,price,brand,camera,os,ram,version):
        super().__init__(price,brand,camera,os,ram)
        self.version=version
        print("Inside AndroidPhone constructor")
a=AndroidPhone(15000,"OnePlus",48,"Android",8,"11")
print(a.version)
```

```python
# Heirarchical Inheritance
# In hierarchical inheritance, multiple child classes inherit from a single parent class. 

class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand= brand
        self.camera=camera 
    def buy(self):
        print("Buying a phone")
        
class SmartPhone(Phone):
    def __init__(self,price,brand,camera,os,ram):
        super().__init__(price,brand,camera)
        self.os=os
        self.ram= ram
        print("Inside the Smartphone constructor")
    def buy(self):
        print("Buying a smartphone")
        super().buy()
        
class FeaturePhone(Phone):
    def __init__(self,price,brand,camera,has_flashlight):
        super().__init__(price,brand,camera)
        self.has_flashlight=has_flashlight
        print("Inside the FeaturePhone constructor")
f=FeaturePhone(5000,"Nokia",2,True)
print(f.has_flashlight)
```

```python
# Multiple Inheritance
# In multiple inheritance, a child class inherits from more than one parent class. This allows the child class to combine the features and behaviors of multiple parent classes.    
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.__price=price
        self.brand= brand
        self.camera=camera 
    def buy(self):
        print("Buying a phone")
        
class Camera:
    def __init__(self,megapixels):
        self.megapixels=megapixels
        print("Inside camera constructor")  
#method  resolution in multiple inheritance
class SmartPhone(Phone,Camera):
    def __init__(self,price,brand,camera,os,ram,megapixels):
        Phone.__init__(self,price,brand,camera)
        Camera.__init__(self,megapixels)
        self.os=os
        self.ram= ram
        print("Inside the Smartphone constructor")
s=SmartPhone(20000,"Samsung",12,"Android",4,48)
print(s.megapixels) 
```

```python
# hybrid Inheritance
# Hybrid inheritance is a combination of two or more types of inheritance. It allows for more complex

class A:
    def method_a(self):
        print("Method A from class A")  
        
class B(A):
    def method_b(self):
        print("Method B from class B")
        
        
        
class C(A):
    def method_c(self):
        print("Method C from class C")

class D(B,C):
    def method_d(self):
        print("Method D from class D")
        
        
d = D()
d.method_a()  # Inherited from class A
d.method_b()  # Inherited from class B
d.method_c()  # Inherited from class C      
d.method_d()  # Defined in class D  
```
