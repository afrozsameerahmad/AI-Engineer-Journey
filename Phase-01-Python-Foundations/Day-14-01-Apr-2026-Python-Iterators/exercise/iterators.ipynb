```python
'''  
what is an  iteration?

Iterator is a general term for taking each item of somthing
one after another .Any time you use a loop explicitly or implicitly to go over a group 
of items ,that is an iteration.
'''

# example
num =[1,2,3,4,5 ]
for i in num:
    print(i)
```

```python
''' 
What is iterable?
Iterable is and object which one can iterate over it 
it generates an iterator when we call iter() on it

'''
# example
l = [1,2,3,4,5]
print(type(l))#l is an iterable
i = iter(l)
print(type(i))#i is an iterator
```
