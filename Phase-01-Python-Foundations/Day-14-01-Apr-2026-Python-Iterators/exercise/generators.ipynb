   #  Genrators
   - Pyhton Generators are simple way to create iterators

```python
# A simple Example
def gen_demo():
    yield 'first statement'
    yield 'second statement'
    yield 'third statement'
g = gen_demo()
for i in g:
    print(i)
```

```python
'''  
Difference between yeild and return statement:
1- return statement is used to return a value from a function and it terminates the function
2- yield statement is used to return a value from a generator function and it does not terminate the function 
it just pauses the function and saves its state for the next time it is called

'''
```

```python
def square_gen(n):
    for i in range(n):
        yield i**2
        
s=square_gen(10)
for i in s:
    print(i)
```

```python
# Range function using generator
def my_range(start,end):
    for i in range(start,end):
        yield i
        
 
for i in my_range(1,10):
    print(i)
```

```python
# Generators expression
g = (i**2 for i in range(10))
print(g)
for i in g:
    print(i)
```

```python
''' 

Benefits of using generators:
1- Memory efficient: Generators do not store all the values in memory at once, they generate values on the fly. This makes them memory efficient and allows us to work with large data sets without running out of memory.
2- ease of implementation: Generators are easy to implement and use. They can be created using simple functions with the yield statement, which makes them a convenient way to create iterators.
3- representing infinite streams of data: Generators can be used to represent infinite streams of data, such as data from a sensor or a network connection. This allows us to process data as it arrives without having to wait for the entire data set to be available.
4-Chaining generators: Generators can be easily chained together to create complex data processing pipelines. This allows us to break down complex problems into smaller, more manageable pieces and makes our code more modular and reusable.
'''

#memory efficient 
l =[x for x in range(10000)]
gen =(x for x in range(10000))

import sys
print(sys.getsizeof(l))
print(sys.getsizeof(gen))

 
```

```python
# representing infinite streams of data
def all_even():
    n = 0
    while True:
        yield n
        n += 2
even_gen = all_even()
print(next(even_gen))
print(next(even_gen))
print(next(even_gen))
```

```python
# chaining generators
def fibonacci_gen():
    a,b = 0,1
    while True:
        yield a
        a,b = b,a+b
fib_gen = fibonacci_gen()
for i in range(10):
    print(next(fib_gen))
```
