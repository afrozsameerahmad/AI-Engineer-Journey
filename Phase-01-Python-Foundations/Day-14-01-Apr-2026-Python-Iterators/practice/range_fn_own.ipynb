```python
class MyRange:
    def __init__(self,start,end):
        self.start = start
        self.end = end
    
    def __iter__(self):
        return MyRange_iterator(self)
```

```python
class MyRange_iterator:
    def __init__(self,iterable_object):
        self.iterable = iterable_object
        
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.iterable.start >= self.iterable.end:
            raise StopIteration
        else:
            current = self.iterable.start
            self.iterable.start += 1
            return current
            
```

```python
x = MyRange(1,10)
print(x)
print(iter(x))
```
