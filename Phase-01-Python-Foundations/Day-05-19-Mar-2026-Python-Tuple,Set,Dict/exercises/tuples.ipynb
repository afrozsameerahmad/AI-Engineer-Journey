#  2- Accessing item in tuple
- Indexing 
- Slicing

```python
'''
Indexing
'''

t=(1,2,3,4,5)
#Positive indexing
print(t[0]) #accessing first element
print(t[1]) #accessing second element

#Negative indexing
print(t[-1]) #accessing last element
print(t[-2]) #accessing second last element

```

```python
'''
Slicing
'''
t=(1,2,3,4,5)
#Positive slicing   
print(t[0:3]) #accessing first three elements
print(t[2:]) #accessing from third element to end
print(t[:3]) #accessing first three elements
#Negative slicing
print(t[-3:]) #accessing last three elements
print(t[:-2]) #accessing all elements except last two
print(t[-4:-1]) #accessing from second element to second last element

#jump slicing
print(t[0:5:2]) #accessing first, third and fifth element
print(t[1:5:2]) #accessing second and fourth element
```

```python
'''
Deleting tuple
we cannot delete individual elements from a tuple because tuples are immutable.
But we can delete the entire tuple using del keyword.

'''
t=(1,2,3)
print(t)
del t
```

```python
'''
Operations on tuple
1. Concatenation
2. Repetition
3. Membership
4. Iteration


'''
l=(1,2,3)
t=(4,5,6)
#concatenation tuple
print(l+t) 
#multiplication tuple
print(l*2)
```

```python
#membership
print(1 in l) 
```

```python
#iteration
for i in l:
    print(i)  
```

```python
# count and index method
t=(1,2,3,4,5,1,2,3)
print(t.count(1)) #count number of times 1 is present in tuple
print(t.index(3)) #index of first occurrence of 3 in tuple
```

```python
''' 
 Difference between list and tuple
 
 
 - SYNTAX - Lists use square brackets [], tuples use parentheses ()
 - MUTABILITY - Lists are mutable, tuples are immutable
 - Speed - Tuples are generally faster than lists
 -Memory - Tuples use less memory than lists    
 - Built in functions : Lists have more built in functions than tuples. For example, lists have append(), extend(), insert(), remove(), pop(), clear() etc. while tuples have only count() and index() methods.
 -error prone : since tuples are immutable, they are less error prone than lists. We cannot accidentally change the values of a tuple, while we can change the values of a list.
 - usability : tuples are used when we want to store a collection of items that should not be changed, while lists are used when we want to store a collection of items that can be changed.

'''
```

```python
# Speed
import time
l1=list(range(1000000))
t1=tuple(range(1000000))
start=time.time()
for i in range(1000000):
    l1[i]
print("Time taken by list: ",time.time()-start)
start=time.time()
for i in range(1000000):
    t1[i]
print("Time taken by tuple: ",time.time()-start)
```
