#  Sets
- A set is a unordered collection of items 
- every set is unique which means no duplicates  and must be immutable (cannot be changed)
- however as set itself is mutable . we can add remove add items in it
- set can also be used to perform mathematical operations like union , intersection, symmetric difference etc 

#  Characteristics 
- unordered
- mutable 
- no duplicates 
- cannot contain mutable data types ---- this point is comfusing lets understand in brief
- in set you cannot have mutable datatype inside set 
- in hindi set khud mutable hota u can delete add but aap uske andr kisi mutable datatypes ko nhi rkh skte 

```python
# Empty set is represented by {} but it is not a set it is a dictionary. To create an empty set we have to use set() function.

set1 = {}
print(type(set1))  
set2 = set()
print(type(set2))  
```

```python
#1d 
set1 = {1, 2, 3, 4, 5}
print(set1)
```

```python
#heterogeneous set
set2 = {1, "Hello", 3.14, True}
print(set2)
# true is considered as 1 in set and it will not be added to the set because sets do not allow duplicate values. 
# So the output will be {1, 'Hello', 3.14}
```

```python
# set using set() function
set3 = set([1, 2, 3, 4, 5])
print(set3)
```

```python
# duplicate values in set not allowed
set4 = {1, 2, 3, 4, 5, 1, 2, 3}
print(set4)
```

```python
# Accessing set elements
set5 = {1, 2, 3, 4, 5}
# sets do not support indexing and slicing because they are unordered collections.
for element in set5:
    print(element)
    
```

```python
# adding elements to set
set6 = {1, 2, 3}
set6.add(4)
print(set6)
#update set with multiple elements
set6.update([5, 6, 7])
print(set6)
```

```python
#deleting elements from set
set = {1, 2, 3, 4, 5}
set.remove(3)
print(set)
# remove() method will raise an error if the element is not found in the set.
```

```python
# deleting elements from set using del keyword
set = {1, 2, 3, 4, 5}
del set
print(set)
```

```python
# discard() method will not raise an error if the element is not found in the set.
set = {1, 2, 3, 4, 5}
set.discard(3)
print(set)
set.discard(6)
print(set)
```

```python
# pop() method removes random element from the set and returns it. If the set is empty, it raises a KeyError.
set = {1, 2, 3, 4, 5}
removed_element = set.pop()
print("Removed element:", removed_element)
print("Set after pop:", set)
```

```python
# clear() method removes all elements from the set, leaving it empty.
set = {1, 2, 3, 4, 5}
set.clear()
print(set)
```

```python
# union of sets
setA = {1, 2, 3}
setB = {3, 4, 5}
union_set = setA.union(setB)
print("Union of setA and setB:", union_set)
# intersection of sets
intersection_set = setA.intersection(setB)
print("Intersection of setA and setB:", intersection_set)
# difference of sets
difference_set = setA.difference(setB)
print("Difference of setA and setB:", difference_set)
# symmetric difference of sets
symmetric_difference_set = setA.symmetric_difference(setB)
print("Symmetric difference of setA and setB:", symmetric_difference_set)   
```

```python
# membership test
setA = {1, 2, 3}
print(2 in setA)  
print(4 in setA)  
```

```python
# iterating through a set
setA = {1, 2, 3, 4, 5}
for element in setA:
    print(element)
```

```python
# frozenset is an immutable version of set. It is created using frozenset() function.
# It does not support add(), remove(), clear() methods because it is immutable.
frozen_set = frozenset([1, 2, 3, 4, 5])
print(frozen_set)
```
