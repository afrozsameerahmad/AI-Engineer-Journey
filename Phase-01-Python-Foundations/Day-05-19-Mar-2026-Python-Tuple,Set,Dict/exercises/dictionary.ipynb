```python
''' 
DICTIONARY

A dictionary in Python is a collection of key-value pairs. Each key is unique and maps to a value. Dictionaries are mutable, meaning you can change their content after they are created. 
They are defined using curly braces {} with key-value pairs separated by colons.
In some other programming languages, dictionaries are known as hash maps or associative arrays.
Example of a dictionary:
my_dict = {'key1': 'value1', 'key2': 'value2', 'key3': 'value3'}

characteristics of dictionary:
mutable: You can change the content of a dictionary after it is created.
indexing has no meaning in dictionary because it is unordered collection of key-value pairs.
keys must be unique and immutable (like strings, numbers, or tuples), while values can be of any data type and can be duplicated.


'''
```

```python
'''
Create dictionary:
'''

#empty dictionary
my_dict = {}
print(my_dict)

#1d dictionary
my_dict = {'name': 'Sameer', 'age': 21, 'city': 'Lucknow'}
print(my_dict)

# with mixed keys
my_dict = {'name': 'Sameer', 1: [1, 2, 3], 'age': 21}
print(my_dict)

#2d dictionary
my_dict = {'name': 'Sameer', 'age': 21, 'address': {'city': 'Lucknow', 'state': 'UP'}}
print(my_dict)

#using sequence and dict() function
my_dict = dict([('name', 'Sameer'), ('age', 21), ('city', 'Lucknow')])
print(my_dict)

#duplicate keys
my_dict = {'name': 'Sameer', 'age': 21, 'name': 'Sameer Ahmad'}
print(my_dict)   
# In this case, the second 'name' key overwrites the first one, so the
# value associated with 'name' in the dictionary will be 'Sameer Ahmad'.
```

```python
# Accessing items from dictionary
my_dict = {'name': 'Sameer', 'age': 21, 'city': 'Lucknow'}
print(my_dict['name'])  
print(my_dict['age'])
print(my_dict['city'])
```

```python
# remove key-value pair from dictionary
my_dict = {'name': 'Sameer', 'age': 21, 'city': 'Lucknow'}
del my_dict['age']
print(my_dict)
```

```python
#pop method removes the specified key and returns the corresponding value. If the key is not found, it raises a KeyError.
#popitem() method removes and returns an arbitrary key-value pair from the dictionary. If the dictionary is empty, it raises a KeyError.

my_dict = {'name': 'Sameer', 'age': 21, 'city': 'Lucknow'}
removed_value = my_dict.pop('age')
print("Removed value:", removed_value)
print("Dictionary after pop:", my_dict)
```

```python
# clear() method removes all items from the dictionary, leaving it empty.
my_dict = {'name': 'Sameer', 'age': 21, 'city': 'Lucknow'}
my_dict.clear()
print(my_dict)

```

```python
# accesing values from 2d dictionary
my_dict = {'name': 'Sameer', 'age': 21, 'address': {'city': 'Lucknow', 'state': 'UP'}}
print(my_dict['address']['city'])
```

```python
# dictionary functions and methods
my_dict = {'name': 'Sameer', 'age': 21, 'city': 'Lucknow'}
print(len(my_dict))   
print(my_dict.keys())   
print(my_dict.values())  
print(my_dict.items())   
```

```python
# dict comprehension
squares = {x: x**2 for x in range(1, 6)}
print(squares)
```
