### How objects access attributes

```python
class Person:

  def __init__(self,name_input,country_input):
    self.name = name_input
    self.country = country_input

  def greet(self):
    if self.country == 'india':
      print('Namaste',self.name)
    else:
      print('Hello',self.name)
```

```python
#How to access attribute of a class
p1 = Person('Sameer', 'India')
print(p1.name)
print(p1.country)   
```

```python
# what if i try to access an attribute that does not exist
print(p1.gender)  
```

### Reference Variables

- Reference variables hold the objects
- We can create objects without reference variable as well
- An object can have multiple reference variables
- Assigning a new reference variable to an existing object does not create a new object

```python
# Object without refrence variable
class Person:
    def __init__(self):
        self.name='Sameer'
        self.gender='male'
Person()
```

```python
p=Person()
''' 
p is refrence variable that store the address of the object created by the class.

in this p is just a variable that store the address of the object created by the class. 
if we do not store the address in a variable then we will not be able to access the attributes of the object.
'''

q=p
#in this q is also a variable that store the address of the object created by the class.
print(q.name)
print(id(p))
print(id(q))
```

```python
q.name='Sameer Ahmad'
print(p.name)
print(q.name)
```

# Pass By Reference

```python
class Person:   
    
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender
        
# outside the class 

def greet(person):
    print(f"Hi my name is {person.name} and I am a {person.gender}")
    
p = Person("Sameer", "male")
greet(p)
```

```python
'''  
COLLECTIONS OF OBJECTS
# we can create a list of objects.
# we can create a list of objects and access the attributes of the objects using the list.






'''
```

```python

#list of objects
class Person:
    def __init__(self   ,name,gender):
        self.name = name    
        self.gender = gender    
p1 = Person("Sameer", "male")
p2 = Person("Ahmad", "male")
p3 = Person("Ayesha", "female")
l=[p1,p2,p3]
        
print(l)
for i in l:
    print(i.name)   
    
# dictionary of objects
d={'p1':p1,'p2':p2,'p3':p3}
print(d)
print(d['p1'].name)
```
