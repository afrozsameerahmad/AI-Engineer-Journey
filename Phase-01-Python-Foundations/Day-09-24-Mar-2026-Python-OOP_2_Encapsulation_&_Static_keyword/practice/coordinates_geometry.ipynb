### Write OOP classes to handle the following scenarios:

- A user can create and view 2D coordinates
- A user can find out the distance between 2 coordinates
- A user can find find the distance of a coordinate from origin
- A user can check if a point lies on a given line
- A user can find the distance between a given 2D point and a given line



```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return'<{}, {}>'.format(self.x, self.y)
    
    def euler_distance(self, other):
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5
    
    def distance_from_origin(self):
        return self.euler_distance(Point(0, 0)) 
    
    
class Line:
    def __init__(self, p1, p2,p3):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        
    def __str__(self):
        return'{} - {} - {}'.format(self.p1, self.p2, self.p3)
    
    def point_on_line(self, point):
        if self.p1*point.x+self.p2*point.y+self.p3==0:
            return "Point is on the line"
        else:
            return "Point is not on the line"
        
    def shortest_distance_from_point(self, point):
        return abs(self.p1*point.x+self.p2*point.y+self.p3)/((self.p1**2+self.p2**2)**0.5)  
```

```python
p1=Point(1, 2)
p2=Point(4, 6)
print(p1)
print(p2)
print(p1.euler_distance(p2))
print(p1.distance_from_origin())
```

```python
L1=Line(p1, p2, Point(7, 8))
print(L1)
```

```python
L1=Line(1,1,-2)
p1=Point(1, 1)
print(L1)
print(p1)
print(L1.point_on_line(p1))
print(L1.shortest_distance_from_point(p1))
```
