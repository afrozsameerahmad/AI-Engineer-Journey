#  Functions are 1st class citiizen

```python
''' 
Functions are datatypes in Python, which means they can be manipulated and used just like any other data type.
functions are first class objects in Python, which means they can be treated like any other object. 
They can be assigned to variables, passed as arguments to other functions, and returned from functions.
 
 
 
 ---type and id 
'''
def square(x):
    return x * x    

print(square(5))
type(square)
```

```python
# reassigning the function to a variable
s = square
print(s(5))
print(type(s))  

# basically we assigned the function to a new variable, and we can call it using that variable.
```

```python
# deleting the original function
del square
print(type(square))  # This will raise an error because 'square' is deleted
 
```

```python
print(s(5))  # This will still work because 's' is a reference to the function
```

```python
# STORING 
l = [1,2,3,4,5,square]
print(l)
```

```python
l[-1](6)  # This will call the function stored in the list and return 36
```

```python
# returing a function from a function
def f():
    def x(a,b):
        return a + b
    return x
val = f()(3,4)
print(val)
# explanation above code in detail
# In the above code, we have a function `f` that defines another function `x` inside it.
# The function `x` takes two arguments `a` and `b`, and returns their sum. 
# The function `f` then returns the function `x` itself, not the result of calling `x`.
# When we call `f()`, it returns the function `x`.
# We can then call the returned function `x` with arguments `3` and `4`, which will return `7`.
```

```python
# functions as arguments to other functions
def func_a():
    print("Hello from func_a")
def func_b(z):
    print("Hello from func_b")
    return z()
    
print(func_b(func_a))  # This will print "Hello from func_b" followed by "Hello from func_a"
```

```python
'''
                   ----LAMBDA FUNCTIONS---
                   
                   
Lambda functions are anonymous functions in Python, which means they do not have a name.
They are defined using the `lambda` keyword and can take any number of arguments, but can only have one expression.
The syntax for a lambda function is:
lambda arguments: expression
Lambda functions are often used for short, simple functions that are not reused elsewhere in the code.
Example of a lambda function:
square = lambda x: x * x
print(square(5))  # Output: 25
In this example, we defined a lambda function that takes one argument `x` and returns the
square of `x`. We then called the lambda function with the argument `5`, which returned `25`.


'''
```

```python
# X-> X**2
square = lambda x: x * x
print(square(5))  
```

```python
'''# difference between normal function and lambda function

# 1. Syntax: Normal functions are defined using the `def` keyword, while lambda
functions are defined using the `lambda` keyword.
# 2. Name: Normal functions have a name, while lambda functions are anonymous and do not have a name.
# 3. Reusability: Normal functions can be reused throughout the code, while lambda 
functions are typically used for short, simple functions that are not reused elsewhere in the code.


Then why we use lambda functions if they are not reusable and do not have a name?

Lambda functions used for higher-order functions,
which are functions that take other functions as arguments or return functions as their result.

'''
```

```python
# Check a string has 'a'
a=lambda s: 'a' in s
a("apple") 
```

```python
b=lambda x:'even' if x%2==0 else 'odd'   
print(b(4))  
```

```python
# High order functions are functions that take other functions as arguments or return functions as their result.

def square(x):
    return x * x

def transform(func, ls):
    output=[]
    for i in ls:
        output.append(func(i))
    print(output)
    
ls = [1,2,3,4,5]
# transform(square, ls)  
transform(lambda x: x ** 2 , ls)  # This will apply the lambda function to each element in the list and print the squared values  


# explanation of above code
# In the above code, we have a function `square` that takes a number `x` and returns its square.
# We also have a function `transform` that takes a function `func` and a list `ls` as arguments.
# The `transform` function iterates through each element in the list `ls`, applies the function `func` to it,
# and appends the result to a new list called `output`.
# Finally, it prints the `output` list.
# We then create a list `ls` with the values `[1, 2, 3, 4, 5]` and call the `transform` function, passing the `square` function and the list `ls` as arguments.
# The `transform` function applies the `square` function to each element in the list `ls`, resulting in a new list `[1, 4, 9, 16, 25]`, which is then printed.  
```

```python
''' 
MAP FUNCTION
The `map` function in Python is a built-in function that applies a given function to each item of an iterable 
(like a list) and returns a map object (which is an iterator).

'''

# Square the items of list
list(map(lambda x: x ** 2, [1,2,3,4,5])) 
```

```python
# odd even labelig of list items
l=[1,2,3,4,5]
list(map(lambda x:'even' if x%2==0 else 'odd' , l))  
```

```python
# fetch names from  a list of dicts

users =[
    {
        'name':'sameer',
        'age':21,
        'gender':'male'
    },
    {
        'name':'azzu',
        'age':25,
        'gender':'male'
    },
    {
        'name':'sara',
        'age':22,
        'gender':'female'
    }
]

list(map(lambda users: (users['name'], users['age']), users))
```

```python
''' 
FILTER FUNCTION

The `filter` function in Python is a built-in function that constructs an iterator from elements of an iterable for which a function returns true. In other words, 
it filters the elements of an iterable based on a specified condition.
'''

#NUMBERS GREATER THAN 5
list(filter(lambda x: x > 5, [1,2,3,4,5,6,7,8,9]))
```

```python
fruits=['apple','orange','pineapple','banana']
list(filter(lambda fruits: fruits.startswith('a'),fruits))
```

```python
''' 
REDUCE FUNCTION


'''
import functools
# SUM OF ALL ITEMS
functools.reduce(lambda x,y: x+y, [1,2,3,4,5])
```

```python
#find min
functools.reduce(lambda x,y: x if x<y else y, [1,2,3,4,5])
```
