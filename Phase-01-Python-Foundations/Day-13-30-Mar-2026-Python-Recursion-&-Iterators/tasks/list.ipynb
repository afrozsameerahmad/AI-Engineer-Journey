```python
''' 
Write a recursive function to print all elements of a list. 
Hint: use list & index as parameters.

 

'''
def print_list(lst, index=0):
    if(index == len(lst)):
        return
    print(lst[index])
    print_list(lst, index+1)
fruits = ['apple', 'banana', 'cherry', 'date', 'elderberry']
print_list(fruits)
```
