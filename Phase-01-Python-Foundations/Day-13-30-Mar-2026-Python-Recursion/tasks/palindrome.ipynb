```python
# Palindrome
''' 
Palidrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward (ignoring spaces, punctuation, and capitalization).
example: "madam"

'''
```

```python
def palindrome(text, original=None):
    if original is None:
        original = text

    if len(text) <= 1:
        print(f"{original} is a palindrome")
    else:
        if text[0] == text[-1]:
            palindrome(text[1:-1], original)
        else:
            print(f"{original} is not a palindrome")

palindrome("madamhh")
```
