```python
''' 
Rabbit Problem:

if 2 newborn rabbits are put in a pen,how mnay rabbits will be in the pen after 1 year
Assuming that:
1- always produce one male and one female offspring
2- can reproduce once every month
3- every couple  can reproduce once they are one month old
4- never die


this is fibnacci sequence problem
'''
```

```python
import time
def fib(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
start = time.time()
print(fib(12))
end = time.time()
print(f"Time taken: {end - start} seconds")
```
