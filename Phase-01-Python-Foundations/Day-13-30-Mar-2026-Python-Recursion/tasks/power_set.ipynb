```python
def power_set(nums):
    result = []
    
    def backtrack(index, current):
        if index == len(nums):
            result.append(current[:])
            return
         
        current.append(nums[index])
        backtrack(index + 1, current)
        
      
        current.pop()
        backtrack(index + 1, current)
    
    backtrack(0, [])
    return result[::-1]

print(power_set([1, 2, 3]))
```
