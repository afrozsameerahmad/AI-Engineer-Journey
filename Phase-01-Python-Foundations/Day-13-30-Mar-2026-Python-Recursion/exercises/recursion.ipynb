```python
'''  
RECURSION -- A function that calls itself is called a recursive function.
IT IS A TECHNIQUE IN WHICH A FUNCTION CALL ITSELF TO SOLVE A PROBLEM.
example: factorial of a number n is the product of all positive integers less than or equal to n.

Call stack: A call stack is a data structure that stores information about the active subroutines of a computer program.
It is used to keep track of function calls and their return values.
When a function is called, a new frame is added to the call stack, and when the function returns, the frame is removed from the stack.

base case: A base case is a condition that allows a recursive function to stop calling itself and return a value.
It is the simplest case that can be solved without recursion.

recursive case: A recursive case is a condition that allows a recursive function to call itself in order to solve a problem.


recurrence relation: A recurrence relation is an equation that defines a sequence of values based on the previous values in the sequence.
'''
```

```python
def factorial(n):
    if (n==0 or n==1):
        return 1        
    else:
        return n * factorial(n-1)
print(factorial(5)) 
```
