```python
def is_even(num):
    """
    This function returns if a given number is odd or even
    input - any valid integer
    output- odd/even
    created on -3/20/2026 
    """
    if type(num)==int:
        if num%2==0:
            return 'even'
        else:
            return 'odd'
    else:
        return ("pagal hai kya .....")
```

```python
# function call 
for i in range (1,11):
    x=is_even(i)
    print(x)
```

```python
# Parameter vs Argument

'''
When we create function that is known as parameter example def is_even(num)  in this num is parameter
and when we use function and give any values then it called arguments 

--Types of Arguments----

1- Default Argument - we give default value in our parameter.


2-Positional Argument- example power(3,2)  default behaviour is called Positional Argument 
jis order me arguments ko aap bhejte ho usi order me assign hote hai jaise power(2,3) to 3  a ko assign hoga and 2 b ko


3- Keyword Argument 
In this we send argument using keyword example power(b=3,a=2) it give 8 

'''
#default argument example
def power(a=2,b=2):
    return a**b

power(2)
#2 assigned to a
```

```python
#positional argument
power(3,2)
```

```python
# keyword argument
power(b=3,a=2)
```

```python
'''
*args and **kwargs are special python keywords that are used to pass the variable length of arguments to a function

*args allow us to pass a variable number of non-keyword arguments to a function

*kwargs 

allow us to pass any number of keyword arguments
keyword arguments mean that they contain a key value pair , like py dict

'''
```

```python
def multiply(*args):
    product=1
    for i in args:
        product=product*i
    return product
```

```python
multiply(1,2,4,5,6,7,3)
```
